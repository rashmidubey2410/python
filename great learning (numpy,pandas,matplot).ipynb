# Numpy mathematics 

```python
# Addition of numpy arrays
```

```python
n1=np.array([10,20])
n2=np.array([40,30])
```

```python
n=np.sum([n1,n2])    #it will add all four element(10+20+40+30=100)
n
```

```python
n=np.sum([n1,n2],axis=0)    #it will add vertical element (10+40=50,20+30=500)
n
```

```python
n=np.sum([n1,n2],axis=1)     #it will add horizontal element (10+20=30,40+30=70)
n
```

```python
# joining numpy array
# vstack()  hstack()   coloum_stack() 
```

```python
n1=np.array([10,20,40])
n2=np.array([40,30,50])
n=np.vstack((n1,n2))
n
```

```python
n1=np.array([10,20,40])
n2=np.array([40,30,50])
n=np.hstack((n1,n2))
n
```

```python
n1=np.array([10,20,40])
n2=np.array([40,30,50])
n=np.column_stack((n1,n2))
n
```

# matplotlib


```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# line plot
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color="green",linewidth=3,linestyle=':')
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="green",linewidth=3,linestyle=':')
plt.plot(x,y2,color="red",linewidth=3,linestyle=':')
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(b=2)
plt.show()
```

```python
# bar plot
```

```python
#create dictionary
student={"sam":50,"bob":70,"don":90,"gop":40}
```

```python
names=student.keys()
names
```

```python
# plt takes only list so convert dic to list

names=list(names)    #names=list(student.keys())
names
```

```python
marks=list(student.values())
marks
```

```python
plt.bar(names,marks,color="yellow")

plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")

plt.show()
```

```python
# horizontal plot
plt.barh(names,marks,color="green")

plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")

plt.show()
```
